```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
plt.axis([0,5, 0, 20])
plt.title('My first plot')
plt.plot([1,2,3,4],[1,4,9,16],'ro')
```

```python
t = np.arange(0,2.5,0.1)
y1 = np.sin(np.pi*t)
y2 = np.sin(np.pi*t+np.pi/2)
y3 = np.sin(np.pi*t-np.pi/2)
plt.plot(t,y1,'r_',t,y2,'g+',t,y3,'yd')
```

```python
t = np.arange(0,2.5,0.1)
y1 = np.sin(np.pi * t)
y2 = np.sin(np.pi * t)
plt.subplot(311)
plt.plot(t,y1,'g-.')
plt.subplot(312)
plt.plot(t,y2,'c--')
```

```python
plt.axis([0,5,0,20])
plt.title('plot')
plt.xlabel('plotx')
plt.ylabel('ploty')
plt.text(1,1, 'first', color='r', fontsize='30')
plt.plot([1,2,3,4],[1,4,9,16],'c*')
plt.plot([1,2,6,4],[1,4,5,1],'y*')
plt.legend(['First series','Second series'])
```

```python
pop = np.random.randint(0,100,100) 
n,bins,patches = plt.hist(pop,bins=20)
```

```python
index = [0,1,2,3,4]
values = [5,7,3,4,6]
plt.bar(index,values)
```

```python
labels = ['Nokia','Samsung','Apple','Lumia']
values = [10,30,45,15]
colors = ['yellow','green','red','cyan']
explode = (0, 0.1, 0, 0)
plt.pie(
    values,
    labels=labels,
    colors=colors, autopct='%1.1f%%',
    shadow=True,
    startangle=90,
    explode=explode
)

plt.axis('equal')
```

```python
from mpl_toolkits.mplot3d import Axes3D
xs = np.random.randint(30,40,100)
ys = np.random.randint(20,30,100)
zs = np.random.randint(10,20,100)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(xs,ys,zs)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
```
